```python
import os
from os.path import join
import numpy as np
import pandas as pd
```

```python
weatherTable = pd.read_csv("PastWeatherData.csv")
weatherTable = weatherTable.dropna().drop(weatherTable.columns[0],axis=1).reset_index(drop=True)
```

```python
morningData = weatherTable.filter(regex='9am')
morningCol = morningData.columns.values.tolist()

afternoonData = weatherTable.filter(regex='3pm')
afternoonCol = afternoonData.columns.values.tolist()

listToFilter = afternoonCol + morningCol

baseData = weatherTable.drop(listToFilter, axis=1)
baseData
```

```python
newMorning = pd.concat([baseData, morningData], axis = 1)
newAfternoon = pd.concat([baseData, afternoonData], axis = 1)
```

```python
numRows = 183

morningSmall = newMorning.groupby('Location').head(numRows).reset_index()
afternoonSmall = newAfternoon.groupby('Location').head(numRows).reset_index()
```

```python
morningSmall['Location'].value_counts()
```

```python
morningSmall.to_csv("MorningWeatherDataset.csv")
afternoonSmall.to_csv("AfternoonWeatherDataset.csv")
```
